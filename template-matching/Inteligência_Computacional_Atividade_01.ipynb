```python
def test_template_matching(folder):
    formas = u.get_files(folder)
    conjunto_formas = u.get_random_images(formas)

    for classe in conjunto_formas.keys():
        print("========================================================")
        print("Classe:", classe)
        print("Template:", conjunto_formas[classe][0])

        for key in conjunto_formas.keys():
            print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
            print("Comparando com a classe:", key)
            print("")
            
            psn, mse = [], []

            for f in conjunto_formas[key][1]:
                psn.append(cmp.pearson_correlation(conjunto_formas[classe][0], f))
                mse.append(cmp.mse(conjunto_formas[classe][0], f))
                # print("Imagem:", f)
                # print("Pearson:", psn[-1])
                # print("MSE:", mse[-1])
                # print("")

            print("MÉDIA:")
            print(f"Pearson: {mean(psn)}\nMSE: {mean(mse)}")
        print("========================================================")
```

```python
test_template_matching("fourShapes/")
```

##### Four-Shapes (Etapa A)

```python
def test_feature_matching(folder, num_parts=3, resize=(60, 60), border_color=(255,255,255)):
    formas = u.get_files(folder)
    conjunto_formas = u.get_random_images(formas)

    for classe in conjunto_formas.keys():
        print("========================================================")
        print("Classe:", classe)
        print("Template:", conjunto_formas[classe][0])
        template = cmp.count_pixels(str(conjunto_formas[classe][0]), num_parts, resize, border_color)

        for key in conjunto_formas.keys():
            print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
            print("Comparando com a classe:", key)
            print("")
            
            psn, mse = [], []

            for f in conjunto_formas[key][1]:
                image = cmp.count_pixels(str(f), num_parts, resize, border_color)
                res = cmp.compare_arrays(template, image)
                psn.append(res[0])
                mse.append(res[1])
                # print("Imagem:", f)
                # # print("Vetor", image)
                # print("Pearson:", psn[-1])
                # print("MSE:", mse[-1])
                # print("")

            print("MÉDIA:")
            print(f"Pearson: {mean(psn)}\nMSE: {mean(mse)}")
        print("========================================================")
```

```python
test_feature_matching("fourShapes/", resize=(200, 200), border_color=(255, 255, 255))
```
